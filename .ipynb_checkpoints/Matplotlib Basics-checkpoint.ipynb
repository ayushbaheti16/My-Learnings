```python
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,5,2]
plt.plot(x,y, label = 'Graph1')
plt.show()
```

```python
x = [1,2,3]
y = [4,5,2]
plt.plot(x,y, label = 'Graph1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('my-graph')
plt.show()
```

```python
x = [1,2,3]
y = [4,5,2]

x2 = [8,7,5]
y2 = [-2,-3,-2]

plt.plot(x,y, label = 'Graph1')
plt.plot(x2, y2, label = 'Graph2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('my-graph')
plt.legend()
plt.show()
```

## Scatter Plot

used to show correlation or some sort of relation between two types of variables. (3D scatter plot can compare three types of variables)


```python
x = [1,2,3,4,5,6,7,8]
y=[21,32,24,5,6,7,53,1]
plt.scatter(x, y, label='scatter-plot', color='k')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.show()
```

## Stack Plots

Stack Plots are required when you need to show a portion of something as a part of some whole entity. 
For example, below, we have used stackplot to track time spent on day to day activities, like , sleeping etc. as a part of whole day!

```python
days = [1,2,3,4,5]

sleeping = [8,7,5,9,9]
eating   = [2,3,4,2,3]
working  = [8,8,9,9,2]
playing  = [6,6,6,4,10]

plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])
plt.xlabel('days')
plt.ylabel('hours')
plt.show()
```

stack plot doesn't allow to show legends. So we will not know what color or area represent to what activity of the day, in above case.

There's a work around for the same. Find below:

```python
plt.plot([],[], color='m', label='Sleeping', linewidth=5)
plt.plot([],[], color='c', label='Eating', linewidth=5)
plt.plot([],[], color='r', label='Working', linewidth=5)
plt.plot([],[], color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])
plt.legend()
plt.plot()
```
